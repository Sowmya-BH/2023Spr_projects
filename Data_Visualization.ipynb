```python
# Loading the data into pandas for easy processing
adds = pd.read_csv("panama_papers.nodes.address.csv", low_memory=False)
adds
```

```python
x = pd.DataFrame(adds['countries'].value_counts()).reset_index()
x.columns = ['countries','count']
x.info()
```

```python
plt.style.use("fivethirtyeight")
@ipywidgets.interact(count=[1000,500,100,1],color=['olive','maroon','green','blue','brown','cyan','dimgrey','salmon','darkseagreen','royalblue','lightgreen'])
def plot_bars(count,color):
    #figure(figsize=(15,7))
    #figure(figsize=(15,7))
    filtered_df = x[x['count'] > count]
    filtered_df.plot(kind='bar', x='countries', y='count',color=color,figsize=[20,5])
    plt.title("Countries with entites who open offshore accounts",fontsize=15,fontweight='bold')
    plt.ylabel("Number of offices",fontsize=15)

    #plt.figure()
    plt.show()
#     plt.bar(x = np.array(x[x['count']>1000].countries), height = np.array(x[x['count']>1000].count), color =color,width = 0.8)
    #plt.show()
```

```python
ents = pd.read_csv("panama_papers.nodes.entity.csv", low_memory=False)
ents
```

```python
ents
```

```python
from dateutil.parser import parse
ents['incorporation_date'] = [parse(d) for d in ents['incorporation_date']]
# print(parse(s2))
# print(parse(s3))
```

```python
ents.info()
```

```python
x_ent = pd.DataFrame(ents['countries'].value_counts()).reset_index()
x_ent.columns = ['countries','count']
x_ent
```

```python
plt.style.use("fivethirtyeight")
@ipywidgets.interact(count=[5000,1000,500,100,1],color=['olive','maroon','green','blue','brown','cyan','dimgrey','salmon','darkseagreen','royalblue','lightgreen'])
def plot_bars(count,color):
    filtered_df_ent = x_ent[x_ent['count'] > count]
    filtered_df_ent.plot(kind='bar', x='countries', y='count',color=color,figsize=[20,5])   
    plt.title("Countries with entites who open offshore accounts",fontsize=15,fontweight='bold')
    plt.ylabel("Number of offshore offices",fontsize=15)
    plt.show()
```

```python
# Convert to datetime
ents["incorporation_date"] = pd.to_datetime(ents["incorporation_date"],
                                           format='%d-%b-%Y',errors='coerce')
```

```python
ents["incorporation_year"]=ents["incorporation_date"].dt.year
```

```python
#Effective Date
ents["inactivation_date"] = pd.to_datetime(ents["inactivation_date"],
                                           format='%d-%b-%Y',errors='coerce')
```

```python
ents["inactivation_year"] = ents["inactivation_date"].dt.year
```

```python
#Effective Date
ents["struck_off_date"] = pd.to_datetime(ents["struck_off_date"],
                                           format='%d-%b-%Y',errors='coerce')
```

```python
ents["struck_off_year"] = ents["struck_off_date"].dt.year
```

```python
ents
```

```python
x_year = pd.DataFrame((ents['incorporation_year'].value_counts()).reset_index())
x_year.columns = ['incorporation_year','count']
x_year
```

```python
plt.style.use("fivethirtyeight")
@ipywidgets.interact(count=[5000,1000,500,100,1],color=['olive','maroon','green','blue','brown','cyan','dimgrey','salmon','darkseagreen','royalblue','lightgreen'])
def plot_bars(count,color):
    filtered_df_ent = x_year[x_year['count'] > count]
    filtered_df_ent.plot(kind='bar', x='incorporation_year', y='count',color=color,figsize=[20,5])   
    plt.title("Yearly Incorporation counts",fontsize=15,fontweight='bold')
    plt.ylabel("Number of offshore offices opened",fontsize=15)
    plt.show()
```

```python
x_year_inactivation = pd.DataFrame((ents['inactivation_year'].value_counts()).reset_index())
x_year_inactivation.columns = ['inactivation_year','count']
x_year_inactivation
```

```python
plt.style.use("fivethirtyeight")
@ipywidgets.interact(count=[5000,1000,500,100,1],color=['olive','maroon','green','blue','brown','cyan','dimgrey','salmon','darkseagreen','royalblue','lightgreen'])
def plot_bars(count,color):
    filtered_df_ent = x_year_inactivation[x_year_inactivation['count'] > count]
    filtered_df_ent.plot(kind='bar', x='inactivation_year', y='count',color=color,figsize=[20,5])   
    plt.title("Yearly Incorporation counts",fontsize=15,fontweight='bold')
    plt.ylabel("Number of offshore offices opened",fontsize=15)
    plt.show()
```

```python
x_year_struck_off = pd.DataFrame((ents['struck_off_year'].value_counts()).reset_index())
x_year_struck_off.columns = ['struck_off_year','count']
x_year_struck_off
```

```python
plt.style.use("fivethirtyeight")
@ipywidgets.interact(count=[5000,1000,500,100,1],color=['olive','maroon','green','blue','brown','cyan','dimgrey','salmon','darkseagreen','royalblue','lightgreen'])
def plot_bars(count,color):
    filtered_df_ent = x_year_struck_off[x_year_struck_off['count'] > count]
    filtered_df_ent.plot(kind='bar', x='struck_off_year', y='count',color=color,figsize=[20,5])   
    plt.title("Yearly struck off counts",fontsize=15,fontweight='bold')
    plt.ylabel("Number of offshore offices striken off",fontsize=15)
    plt.show()
```
