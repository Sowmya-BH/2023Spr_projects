```python
# create a bar plot function
def plot_bars(count):
    filtered_df = x[x['count'] > count]
    filtered_df.plot(kind='bar', x='countries', y='count')
    plt.show()


# create an interactive slider widget
count_slider = widgets.IntSlider(value=1000, min=0, max=20267, step=500)

# create a callback function for the slider widget
def on_value_change(change):
    plot_bars(change.new)

# link the slider widget to the callback function
count_slider.observe(on_value_change, names='value')

# display the slider widget
display(count_slider)
```

```python
edges = pd.read_csv("panama_papers.edges.csv", low_memory=False)
edges
```

```python
#create a directed graph using the networkx package
G = nx.DiGraph()

nodes = [(adds, "address"), (ents, "entities"), (inter, "intermediates"), (offi, "officers")]

#adds nodes based on data stored in several Pandas DataFrames
for df, node_type in nodes:
    #data = df.to_numpy()
    for row in df.itertuples(index=False):
        G.add_node(row.node_id, node_type=node_type, details=row._asdict())
        
#adds edges to the graph based on data stored in edges DataFrames
for row in edges.itertuples(index=False):
    G.add_edge(row.START_ID, row.END_ID, rel_type=row.TYPE, details=row._asdict())
```
